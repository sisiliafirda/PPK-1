### 1. Cari bagaimana cara menampilkan Judul(tittle) dari grafik

```python
import numpy as np
import matplotlib.pyplot as plt

x=[1,2,3,4,5,6]
y=[54,52,50,51,49,48]
plt.plot(x,y)

plt.title('Program Penurunan Berat Badan')

plt.grid(True)
plt.show()
```

### 2. Cari bagaimana cara memberi label pada sumbu x dan sumbu y

```python
import numpy as np
import matplotlib.pyplot as plt

x=[1,2,3,4,5,6]
y=[54,52,50,51,49,48]
plt.plot(x,y)

plt.title('Program Penurunan Berat Badan')
plt.xlabel('Pekan')
plt.ylabel('Berat Badan')

plt.grid(True)
plt.show()
```

### 3. Cari bagaimana cara menambahkan teks dalam grafik

```python
import numpy as np
import matplotlib.pyplot as plt

x=[1,2,3,4,5,6]
y=[54,52,50,51,49,48]
plt.plot(x,y)

plt.title('Program Penurunan Berat Badan')
plt.xlabel('Pekan')
plt.ylabel('Berat Badan')
plt.text(4,51,"Berat Badan Mengalami Kenaikkan pada Pekan 4")

plt.grid(True)
plt.show()
```
